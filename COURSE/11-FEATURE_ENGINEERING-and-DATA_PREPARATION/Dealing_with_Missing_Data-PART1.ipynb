### PART 1 - MISSING DATA - EVALUATING of MISSING DATA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
with open('Ames_Housing_Feature_Description.txt','r') as f:
    print(f.read())
```

```python
df = pd.read_csv('Ames_outliers_removed-ORIG.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.drop("PID",axis=1)
```

```python
len(df.columns)
```

```python
100 * df.isnull().sum() / len(df)
```

```python
def percent_missing(df):
    percent_nan = 100 * df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()

    return percent_nan
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
plt.figure(figsize=(8,4))

# Cool to warm gradient
viridis = plt.cm.viridis(np.linspace(0, 1, len(percent_nan.index)))

sns.barplot(x=percent_nan.index,y=percent_nan,palette=list(viridis),hue=percent_nan.index)
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
# BSMT NUMERIC COLUMNS --> fillna to 0
bsmt_num_cols = ['BsmtFin SF 1','BsmtFin SF 2','Bsmt Unf SF','Total Bsmt SF','Bsmt Full Bath','Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)

# BSMT STRING COLUMNS
bsmt_str_cols =  ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_str_cols] = df[bsmt_str_cols].fillna('None')
```
