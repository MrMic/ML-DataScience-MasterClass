## Matplotlib - Exercises / Solutions

### PART 1

```python
import numpy as np

m = np.linspace(0,10,11)
```

```python
m
```

```python
c = 3 * 10**8
```

```python
E = m * c**2
```

```python
E
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(m,E,color='red',lw=2)
plt.title('E = c^2')
plt.xlabel('M in g')
plt.ylabel('E in J')
plt.xlim(0,10)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(m,E,color='red',lw=3)
```
