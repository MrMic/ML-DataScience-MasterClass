## PART 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label="X vs X")
ax.plot(x,x**2,label='X vbs X^2')

ax.legend(loc=(01.1,0.5))
```

## PART 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Possible line Style: '--', '-', '-.', ':', 'steps'
# See also matplotlib.markers

ax.plot(x,x,color='#E4973E',label='Y = X',lw=3,ls='--',marker='s',ms=6,
       markerfacecolor='red',markeredgewidth=3,markeredgecolor='orange') # RGBHex Color

ax.plot(x,x+1,label='Y = X+1',ls='-.',marker='o')

lines = ax.plot(x,x+2,label='Y = X+2',color='green',marker='^',ms=10)
lines[0].set_dashes([1,2,1,2,10,1])

ax.legend()
```
