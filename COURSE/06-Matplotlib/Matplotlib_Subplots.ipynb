```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_xlabel('X LABEL 0,0')
axes[0][0].set_ylabel('Y LABEL 0,0')
axes[0][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title("TITLE 1,1")

fig.suptitle("Figure Level")
plt.tight_layout()
## fig.subplots_adjust(wspace=0.9)
fig.set_figwidth(10)

fig.savefig("new_subplots.png",bbox_inches='tight')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
    ax.plot(x,y)
plt.tight_layout()
```
