```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../DATA/tips.csv")
```

```python
df.head(3)
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```
