# LINEAR REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
```

```python
sns.regplot(data=df,x='total_spend',y='sales')
```

```python
X = df['total_spend']
y = df['sales']
```

```python
# y = mx + b
# y = B1x + B0
# help(np.polyfit)
```

```python
np.polyfit(x=X,y=y,deg=1)
```

```python
potential_spend = np.linspace(1,500,100)
```

```python
preditec_sales = 0.04868788*potential_spend + 4.24302822
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
plt.plot(potential_spend, preditec_sales,color='red')
```
