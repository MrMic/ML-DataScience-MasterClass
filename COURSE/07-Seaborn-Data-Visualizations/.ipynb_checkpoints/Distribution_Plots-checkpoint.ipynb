```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
from summarytools import dfSummary
```

```python
df = pd.read_csv('../../05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
dfSummary(df)
```

```python
plt.figure(figsize=(10,6))
sns.rugplot(data=df,x='salary',height=0.5)
```

```python
# DO NOT USE distplot() => Deprecated
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20,color='red',edgecolor='orange',kde=True,rug=True)
```

```python
sns.set(style='white') # DEFAULT
sns.histplot(data=df,x='salary',kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
samples_ages = np.random.randint(0,100,200)
```

```python
samples_ages
```

```python
samples_ages = pd.DataFrame(samples_ages,columns=['age'])
```

```python
sns.rugplot(data=samples_ages,x='age')
```

```python
sns.displot(data=samples_ages,x='age',rug=True,bins=30,kde=True)
```

```python
sns.kdeplot(data=samples_ages,x='age',clip=[0,100],bw_adjust=0.5,fill=True)
```
