```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('application_record.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
employed = df[df['DAYS_EMPLOYED']<0]
```

```python
employed['DAYS_EMPLOYED'] = employed['DAYS_EMPLOYED']*-1
```

```python
employed['DAYS_BIRTH'] = employed['DAYS_BIRTH']*-1
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=employed,x='DAYS_BIRTH',y='DAYS_EMPLOYED',alpha=0.01,linewidth=0)

plt.savefig('task_one.jpg')
```

```python
df['YEARS'] = -1*df['DAYS_BIRTH']/365

sns.histplot(data=df,x='YEARS',bins=45)
```

```python
bottom_half_income = df.nsmallest(n=int(len(df)/2),columns='AMT_INCOME_TOTAL')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=bottom_half_income,x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',hue='FLAG_OWN_REALTY')
plt.legend(loc=(1.05,0.5),title='FLAG_OWN_REALTY')
plt.title('Income Totals per Family Status for Bottom Half of Earners')
```
