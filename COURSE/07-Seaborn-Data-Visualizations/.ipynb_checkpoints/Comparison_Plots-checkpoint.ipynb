```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from summarytools import dfSummary
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
dfSummary(df)
```

```python
plt.figure(figsize=(10,4))

sns.jointplot(data=df,x='math score',y='reading score',kind='hex')
sns.jointplot(data=df,x='math score',y='reading score',kind='scatter',hue='gender')
sns.jointplot(data=df,x='math score',y='reading score',kind='hist',hue='gender')
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',hue='gender')
```

```python
sns.pairplot(data=df,hue='gender')
```
