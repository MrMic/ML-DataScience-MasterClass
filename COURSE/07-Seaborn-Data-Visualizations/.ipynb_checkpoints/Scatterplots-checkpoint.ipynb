```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=100)
sns.scatterplot(data=df,x='salary',y='sales',hue='level of education',palette='Dark2')
```

```python
plt.figure(figsize=(12,4),dpi=100)
sns.scatterplot(data=df,x='salary',y='sales',size='salary')
```

```python
plt.figure(figsize=(12,4),dpi=100)
sns.scatterplot(data=df,x='salary',y='sales',s=200,alpha=0.3)
```

```python
plt.figure(figsize=(12,4),dpi=100)
sns.scatterplot(data=df,x='salary',y='sales',s=200,hue='level of education',style='level of education')

plt.savefig("my_plot.jpg")
```
