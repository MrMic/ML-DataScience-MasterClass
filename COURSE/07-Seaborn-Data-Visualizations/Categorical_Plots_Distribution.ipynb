```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
from summarytools import dfSummary
dfSummary(df)
```

```python
plt.figure(figsize=(10,4),)
sns.boxplot(data=df,y='reading score',x='parental level of education',hue='test preparation course',palette='husl')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,4),)
sns.violinplot(data=df,y='reading score',x='parental level of education',hue='test preparation course',palette='husl',split=True)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor=(1.20,0.5))
```

```python
sns.boxenplot(data=df,x='math score',y='test preparation course',hue='gender')
plt.legend(bbox_to_anchor=(1.30,0.5))
```
